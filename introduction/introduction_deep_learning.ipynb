# Evaluation de l'erreur du medèle

x = nombre de chambre 
y = prix

```python
def model(x, w):
    return w * x 
```

```python
model(x=3, w=5)
```

```python
def erreur(x, w, y):
    y_model = model(x, w)
    distance = (y -y_model) ** 2

    return np.sum(distance)
```

```python
possible_w = np.arange(-20, 20, 0.001)

possible_w
```

```python
possible_w.size
```

```python
errors = []

for d in possible_w:
    e = erreur(x=n_chambre, w=d, y=prix)
    errors.append(e)
```

```python
plt.plot(possible_w, errors)
plt.xlabel('les w possible')
plt.ylabel('erreurs')
plt.show()
```
